```python
mpl.rcParams.update({
    'font.size'           : 14.0,
    'axes.titlesize'      : 'large',
    'axes.labelsize'      : 'medium',
    'xtick.labelsize'     : 'medium',
    'ytick.labelsize'     : 'medium',
    'legend.fontsize'     : 'large',
})


fig, (ax0, ax1, ax2) = plt.subplots(ncols=3, figsize=(15, 5))

ax0.hist(df['log_aland_sqmi'], 20, facecolor='green', alpha=0.75)
ax0.set_title('County Areas')
ax0.set_ylabel('number of counties')
ax0.set_xlabel('Log of the area (in log(square miles)')

ax1.hist(df['log_pop_est_2015'], 20, facecolor='blue', alpha=0.75)
ax1.set_title('2015 Estimated Populations')
ax1.set_xlabel('Log number of residents')

ax2.hist(df['log_pop_den_2015'], 20, facecolor='red', alpha=0.75)
ax2.set_title('2015 Estimated Population Densities')
ax2.set_xlabel('Log of population density (in log(habitants/square miles))')

plt.savefig('../figs/area_pop_dens_hist.png')
fig.tight_layout()
```

```python
fig = plt.figure(figsize=(8, 5))

ax0 = fig.add_subplot(1,1,1)
ax0.hist(df['pop_ch_10000_2015'], 20, facecolor='green', alpha=0.75)
ax0.set_title('2015 1-year population change')
ax0.set_ylabel('number of counties')
ax0.set_xlabel('Population change for every 10000 people')
fig.tight_layout()
```
